```python
import make_labelled_custom as mlc
from datetime import datetime
import matplotlib.pyplot as plt
import pickle
import numpy as np
dir_flacs = '/2017/wk15' #select week
lst = mlc.get_flac_files('/scratch/dale_data/UK-DALE-2017/UK-DALE-2017-16kHz/house_1'+dir_flacs)

# loads a model. I will try to send you the pickle file but it is around 1.5 Gb
# so this code won't run yet. You can see the generated graphs below
infile = open('forest001.pickle', 'rb')
forest1 = pickle.load(infile)
infile.close()
```

```python
from pysegments._segments import Interval, segment

def vote(interval):
    # finds average prediction over a whole interval
    l, r = interval.inf, interval.sup
    t = timestamps
    ry = yP[np.logical_and(t > l, t < r)]
    return ry.mean()

def thresh(interval):
    # returns true for a threshold
    return vote(interval) > 0.8

def consensus(interval):
    # split in 10 intervals and checks if they give agreeing predictions
    l, r = interval.inf, interval.sup
    a = (np.arange(11)/10) * (r - l) + l
    subintervals = [(Interval(a[i],a[i+1])) for i in range(10)]
    average = vote(interval)
    
    # checks if all intervals give a similar value from the average
    agree = True
    tolerance = 0.1 #0.05 works too
    for s in subintervals:
        if abs(vote(s) - average) > tolerance: 
            agree = False
            break
            
    # if subintervals agree we check wether the average is larger than a threshold
    # one variant would be to return True when even there is no agreement
    return agree and (average > 0.4) #0.15 is nice 0.4 is great but gets a few false negatives too
    
def containing(segments, ts):
    # checks wether a timestamp is in one of the intervals in the list 'segments'
    for interval in segments:
        if interval.contains(ts):
            return 1
    return 0

```

```python
'''The first two graphs are generated by the segmentation method with the two char functions. The third is 
the actual data.'''

def split(data):
    # for getting the times and labels from the data
    return data[:,:-1], data[:,-1]

for fn_flac in lst[:50]: #lst[150:] is a good place to start the loop too
    #read data
    print(fn_flac)
    #in make labelled custom make sure you set debug = True
    copy1 = mlc.get_labelled_flac_file(fn_flac, dir_flacs, logsig=True, convert_va=True)
    aux, y1 = split(copy1) #y1 are the labels
    y1 = 1 - y1 #they are genererated with the full data method so they are flipped
    X1, timestamps = split(aux) 
    
    #predict cycles
    forest = forest1
    yP = forest.predict(X1)
    print(forest.score(X1,y1))
    
    base = Interval(timestamps[0], timestamps[len(timestamps)-1])
    
    
    print("Segmentations in progress..")
    fig = plt.figure()
    segments = segment(base, consensus, 0, 2)
    plt.plot(timestamps, [containing(segments, ts) for ts in timestamps])
    
    fig = plt.figure()
    segments = segment(base, thresh, 0, 2)
    plt.plot(timestamps, [containing(segments, ts) for ts in timestamps])
    
    fig = plt.figure()   
    plt.plot(timestamps,y1)
    plt.show()
```
